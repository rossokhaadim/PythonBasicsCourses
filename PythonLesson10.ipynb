> Треба створити функцію котра приймає текст(story) як аргумент, та повертає СЛОВНИК в котрому ключі це унікальні слова з тексту, а їх значення - кількість входження слова в текст. Тобто скільки разів у тексті зустрічається те чи інше слово.



```python
def dic(texxt):
  word_dict={}
  s=''
  x=1
  for i, word in enumerate(texxt):
    if word == '.' or word == ',' or word == '?' or word == '!' or word == '"' or word == "'":
      texxt = texxt.replace(word, " ")
  texxt = sorted(texxt.split())
  word_dict={}
  x=1
  for i, te in enumerate(texxt):
    if(i!=len(texxt)-1):
      if(te == texxt[i+1]):
        x+=1
      else:
        word_dict.update({te : x})
        x=1
    else:
      word_dict.update({te : x})
  return word_dict
```

## Завдання 6


> Створіть функцію котра отримує 1 ім'я writers або director та поверне словник з іменем сценариста або режисера, кількіст його фільмів, місце в рейтингу та назви цих фільмів зі змінної

```python
def search1(name):
  actor_name = {}
  film_num = 0
  rating_list = []
  film_names = []
  for film in imdb_rank:
    if((name in film['director']) or (name in film['writers'])):
      film_num+=1
      rating_list.append(film['rating'])
      film_names.append(film['title'])
  ave_raiting = 0
  for i in rating_list:
    ave_raiting+=float(i)
  ave_rating = ave_raiting/len(rating_list)
  actor_name.update({'Name': name})
  actor_name.update({'Number of films': film_num})
  actor_name.update({'Rating': rating_list})
  actor_name.update({'Average rating': ave_rating})
  actor_name.update({'Films titles': film_names})
  return actor_name
```
