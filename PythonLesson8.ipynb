## Завдання 2


> Використання циклу while, операторb if і функції str(); ітеруй список, якщо знайшов 100, надрукуй його разом з номером індексу: "Знайшов 100 під індексом №: 5". Після знаходження шуканого числа, вийди з цикла за допомогою break



````python
lst = [10, 99, 98, 85, 45, 59, 65, 66, 76, 12, 35, 13, 100, 80, 95, 12, 255, 124, 466, 9]
````

````python
Condition = True
text1 = 'Знайшов '
text3 = ' під індексом №:'
while Condition:
  for k, n in enumerate(lst):
    if n == 100:
      text2 = str(n)
      text4 = str(k)
      text = text1 + text2 + text3 + text4
      print(text)
      Condition = False
      break
    else:
      k+=1
````

# Робота з GIT

## Налаштування GIT




## Встановлюємо git



> [Встановлюємо git](https://git-scm.com/)


### Крок 1


Створіть аккаунт на GitHub



> [github](https://github.com/signup)


### Крок 2

Створіть репозиторій, назва hillel_git



> [Create new repository](https://github.com/new)





### Крок 3

В PyCharm створіть новий проект(або використайте вже створенний) з віртуальним оточенням у папці venv та 1 файлом main.py у корні проекту

### Крок 4

Введіть глобальні налаштування git на вашому комп'ютері, в терміналі введіть по черзі:



```
git config --global user.name "Your Name"
git config --global user.email "your mail@sopulit.com"
```



### Крок 5

Створіть локальий репозиторій у вашому проекті, в терміналі команда:

`git init`

### Крок 6

Додайте всі файли на трекінг git, в терміналі команда(крапка в команді обов'язкова):
`git add .`


### Крок 7

Проіндексуйте поточний стан файлів проект(знімок проект), в терміналі комнда:
`git commit -m "first commit"`

### Крок 8

Звяжіть локальний проект з репозиторієм на GitHub, в терміналі комнда:


```
git remote add origin https://github.com/YOUR_OWN_REPOSITORY_URL
```



### Крок 9

Вигрузіть ваш локальний проект в репозиторій, в терміналі комнда:

`git push -u origin main`

або

`git push -u origin master`
