# Завдання 1



> Ситворіть функцію, котра приймає список чисел та повертає тупл. Перше значення тупла є середнім значенням всіх аргументів, а друге значення перше число яке більше середнього.




```python
def function(listt):
  ave = sum(listt)/len(listt)
  for test in listt:
    if test > ave:
      bigger = test
      break
  list_tuple = (ave, bigger)
  return list_tuple

test_list = [1, 2, 3, 4, 12, 322, 55, 66, 71, 80, 94, 6, 10, 32]
calculations = function(test_list)
calculations
```

```python
def pair(listk):
  listtt = [ ]
  for value in listk:
    if value % 2 == 0:
      listtt.append(value)
  return listtt
new_list = pair(test_list)
print(new_list)
```

# Завдання 3

> Створіть функцію, яка приймає список туплів і повертає назву предмету з найвищою оцінкою.

```python
def highest(gradees):
  max = 0
  subject = ''
  for gradee in gradees:
    if gradee[1] > max:
      max = gradee[1]
      subject = gradee[0]
  return subject
grade = [('Англійська мова', 88), ('Біологія', 90), ('Математика', 97), ('Українська мова', 82)]
sub = highest(grade)
sub
```
